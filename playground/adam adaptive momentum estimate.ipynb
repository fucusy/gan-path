```lua
dw = torch.randn(10, 1)
w = torch.randn(10, 1)
```

```lua
beta1 = 0.9
beta2 = 0.999
eps = 10e-8
lr = 10e-3
m = torch.zeros(dw:size())
v = torch.zeros(dw:size())
```

```lua
for t=1, 10 do
    m:mul(beta1):add(torch.mul(dw, 1-beta1))
    v:mul(beta2):add(torch.mul(torch.pow(dw, 2), 1-beta2))
    
    -- correct m and v
    m:div(1-math.pow(beta1, t))
    v:div(1-math.pow(beta2, t))
    
    v:sqrt():add(eps)
    m:mul(-1*lr)
    w:add(m:cdiv(v))
    print("m", m)
    print("v", v)
    print("w", w)
end
```
